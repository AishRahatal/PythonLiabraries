```python
'''
@Author: Aishwarya
@Date: 2021-12-7
@Title: plot two or more lines with different styles

'''
########################################################################################################
import matplotlib.pyplot as plt

x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [10,30,0]

x3 = [10,20,30]
y3 = [30,50,20]


x4= [10,20,27]
y4 = [5,20,0]
plt.xlabel('x - axis')


plt.title('Two or more lines with different widths and colors with suitable legends ')

plt.plot(x1,y1, color='indigo',  label = 'line1-dotted',ls='dotted')
plt.plot(x2,y2, color='blue', label = 'line2-dashed',ls='dashed')
plt.plot(x3,y3, color='grey', label = 'line3-dashdot',linestyle='dashdot')
plt.plot(x4,y4, color='violet', label = 'line4-solid',linestyle='solid')
# show a legend on the plot
plt.legend()
plt.show()
```
