```python
'''
@Author: Aishwarya
@Date: 2021-12-8
@Title: plot two or more lines with different markers

'''
########################################################################################################
import matplotlib.pyplot as plt
# x axis values
x = [1,4,5,6,7]
# y axis values
y = [2,6,3,6,3]

# plotting the points 
plt.plot(x, y, color='black', linestyle='dashed', linewidth = 2,marker='o', markerfacecolor='black', markersize=12)
#Set the y-limits of the current axes.
plt.ylim(1,8)
#Set the x-limits of the current axes.
plt.xlim(1,8)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('Marker')
# function to show the plot
plt.show()
```
