```python
'''
@Author: Aishwarya
@Date: 2021-12-7
@Title:draw a line with suitable label in the x axis, y axis and a title using data from text file
'''
########################################################################################################
import matplotlib.pyplot as plt
with open("test.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
plt.plot(x, y)

plt.xlabel('x - axis')

plt.ylabel('y - axis')
 
plt.title('Straight line')

plt.show()
```
