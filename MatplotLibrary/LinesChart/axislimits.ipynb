```python
'''
@Author: Aishwarya
@Date: 2021-12-8
@Title: Display lines axis limits

'''
########################################################################################################
import matplotlib.pyplot as plt
# x axis values
x = [1,4,5,6,7]
# y axis values
y = [2,6,3,6,3]

# plotting the points 
plt.plot(x, y,linewidth = 2)
#Set the y-limits of the current axes.
plt.ylim(1,8)
#Set the x-limits of the current axes.
plt.xlim(1,8)

# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')

# giving a title to my graph
plt.title('Axis Limits x-y=8')
# function to show the plot
plt.show()
```
