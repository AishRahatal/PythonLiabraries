```python
'''
@Author: Aishwarya
@Date: 2021-12-9
@Title: draw a scatter plot of “day” against “total bill” for a dataset
'''
########################################################################################################
import matplotlib.pyplot as plt
import seaborn as s

# Set Seaborn style
s.set_style('darkgrid')
# Import Data
tips = s.load_dataset('tips')
print(tips)
# Construct plot
s.scatterplot(x = "day", y = "total_bill", data = tips)
plt.show()
```
