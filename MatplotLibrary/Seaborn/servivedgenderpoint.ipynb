```python
'''
@Author: Aishwarya
@Date: 2021-12-9
@Title: draw point plot of sex against survived for a dataset
'''
########################################################################################################
import matplotlib.pyplot as plt
import seaborn as s

# Set Seaborn style
s.set_style('darkgrid')
# Import Data
titanic_dataset = s.load_dataset("titanic")
print(titanic_dataset)
# Construct plot
s.pointplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```
