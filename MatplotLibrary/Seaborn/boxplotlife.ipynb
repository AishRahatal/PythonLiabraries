```python
'''
@Author: Aishwarya
@Date: 2021-12-9
@Title: draw box plot of life expectancy by continent for a data set
'''
########################################################################################################
import matplotlib.pyplot as plt
import seaborn as s
import pandas as pd

# Set Seaborn style
s.set_style('darkgrid')
# Import Data
df = pd.read_csv('year.csv')
print(df)
# Construct plot
plt.figure(figsize=(50, 50))
s.boxplot(x = df['country'], y = "lifeExp", data = df)#bubble plot
plt.show()
```
