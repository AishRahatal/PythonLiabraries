```python
'''
@Author: Aishwarya
@Date: 2021-12-9
@Title: draw a violin plot of sex against day for a given dataset
'''
########################################################################################################
import matplotlib.pyplot as plt
import seaborn as s

# Set Seaborn style
s.set_style('darkgrid')
# Import Data
tips = s.load_dataset('tips')
print(tips)
# Construct plot

ax = s.violinplot(x="day", y="total_bill", hue="sex",

                    data=tips, palette="Set2", split=True,

                    scale="count", inner="stick",

                    scale_hue=False, bw=.2)
plt.show()
```
