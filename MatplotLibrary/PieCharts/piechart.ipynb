```python
'''	
@Author: Aishwarya
@Date: 2021-12-8
@Title : create a pie chart of the popularity of programming
Languages.
'''
########################################################################################################
import matplotlib.pyplot as plt
# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)  
# Plot
fig = plt.figure()
fig.patch.set_facecolor('white')
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,autopct='%1.1f%%', shadow=False, startangle=140)

plt.axis('equal')
plt.show()

```
